```python
import cv2
import matplotlib.pyplot as plt
import requests
from io import BytesIO

def url_to_image(url):
    response = requests.get(url)
    image_bytes = BytesIO(response.content)
    image = cv2.imdecode(np.frombuffer(image_bytes.read(), np.uint8), -1)
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

def red_filter(image):
    filtered_photo = np.copy(image)
    filtered_photo[:,:,1:] = 0
    return filtered_photo

def green_filter(image):
    filtered_photo = np.copy(image)
    filtered_photo[:,:,0] = 0
    filtered_photo[:,:,2] = 0
    return filtered_photo

def blue_filter(image):
    filtered_photo = np.copy(image)
    filtered_photo[:, :, :2] = 0
    return filtered_photo

def redandgreen_filter(image):
    filtered_photo = np.copy(image)
    filtered_photo[:,:,2] = 0
    return filtered_photo

def redandblue_filter(image):
    filtered_photo = np.copy(image)
    filtered_photo[:,:,1] = 0
    return filtered_photo

def greenandblue_filter(image):
    filtered_photo = np.copy(image)
    filtered_photo[:,:,0] = 0
    return filtered_photo

def apply_filters(image):
    red_filtered_photo = red_filter(image)
    green_filtered_photo = green_filter(image)
    blue_filtered_photo = blue_filter(image)
    red_and_green_filtered_photo = redandgreen_filter(image)
    red_and_blue_filtered_photo = redandblue_filter(image)
    green_and_blue_filtered_photo = greenandblue_filter(image)
    return red_filtered_photo, green_filtered_photo, blue_filtered_photo, red_and_green_filtered_photo, red_and_blue_filtered_photo, green_and_blue_filtered_photo

def display_images(image):
    original_image = image
    red_filtered_photo, green_filtered_photo, blue_filtered_photo, red_and_green_filtered_photo, red_and_blue_filtered_photo, green_and_blue_filtered_photo = apply_filters(image)

    plt.figure(figsize=(15, 10))

    plt.subplot(2, 4, 1)
    plt.imshow(original_image)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(2, 4, 2)
    plt.imshow(red_filtered_photo)
    plt.title('Red Filtered Photo')
    plt.axis('off')

    plt.subplot(2, 4, 3)
    plt.imshow(green_filtered_photo)
    plt.title('Green Filtered Photo')
    plt.axis('off')

    plt.subplot(2, 4, 4)
    plt.imshow(blue_filtered_photo)
    plt.title('Blue Filtered Photo')
    plt.axis('off')

    plt.subplot(2, 4, 6)
    plt.imshow(red_and_green_filtered_photo)
    plt.title('Red and Green Filtered Photo')
    plt.axis('off')

    plt.subplot(2, 4, 7)
    plt.imshow(red_and_blue_filtered_photo)
    plt.title('Red and Blue Filtered Photo')
    plt.axis('off')

    plt.subplot(2, 4, 8)
    plt.imshow(green_and_blue_filtered_photo)
    plt.title('Green and Blue Filtered Photo')
    plt.axis('off')

    plt.show()

url = input("Please enter the URL of a .jpg file: ")
image = url_to_image(url)
display_images(image)
```
