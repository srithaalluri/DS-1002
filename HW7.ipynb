```python
messy_df = pd.read_csv('/content/more-messy-data.csv')

messy_df
```

```python
print(messy_df.duplicated().sum())
```

```python
messy_df = messy_df.drop_duplicates()

messy_df.duplicated().sum()
```

```python
messy_df.info()
for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    messy_df[col] = pd.to_numeric(messy_df[col], errors='coerce')
for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    messy_df[col] = messy_df[col].fillna(messy_df[col].mean())
print(messy_df.info())
```

```python
messy_df.isnull().sum()
# this code chunk shows that there aren't any missing values now
```

```python
messy_df['species'].value_counts()
# there are 10 species
```

```python
messy_df['species'].value_counts()

messy_df['species'] = messy_df['species'].str.lower()

messy_df['species'] = messy_df['species'].str.replace('setoso', 'setosa')
messy_df['species'] = messy_df['species'].str.replace('setosan', 'setosa')
messy_df['species'] = messy_df['species'].str.replace('virgininca', 'virginica')
messy_df['species'] = messy_df['species'].str.replace('virginia', 'virginica')
messy_df['species'] = messy_df['species'].str.replace('viginica', 'virginica')


print(messy_df['species'].value_counts())
# output below shows the 3 values for the 'species' variable
```
