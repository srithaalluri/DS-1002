13.  Among the four variables, the two that are most negatively correlated (overall) are sepal_width and petal_length. (How do we know this?) Determine the same thing but for each species separately: for each species, which two variables have the most negative (or the least positive) correlation? Again, include calculations to support your answer.

```python
sepal_data = iris[['species', 'sepal_length', 'sepal_width']]

species_correlation = sepal_data.groupby('species').corr()

species_correlation_sepal = species_correlation.iloc[0::2, 1]

print(species_correlation_sepal)

# The following code chunk between 'sepal_length' and 'sepal_width' for each of the species separately
# The least positive correlations belonged to the following variables: 'virginica' and 'versicolor'
```

14.  Once again, the variables `sepal_width` and `petal_length` are negatively correlated.  But for each individual species, these two variables are positively correlated.  How is this possible?  Explain this phenomenon.  Use a data visualization if it helps.


```python
print(iris.plot.scatter('petal_length','sepal_width'))

print(sns.relplot(iris, x='sepal_length', y='petal_length', hue='species'))
```

Although each individual species has a positive correlation for both variables, they vary in their strength. For example, versicolor and virginica are the two species with weaker positive correlations. Therefore, we observe an overall negative correlation when examining petal length and sepal width with combined species.
